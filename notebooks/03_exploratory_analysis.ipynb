Branch Summary

```python
branch_summary = (
    master_loans.groupby("shop")
    .agg(
        Phones=("agreement_number", "count"),
        Loan_Value=("loan_amount", "sum"),
        Outstanding_Debt=("debt", "sum"),
        Average_DPD=("dpd", "mean")
    )
    .reset_index()
)

branch_summary.head()
```

```python
branch_summary.to_csv(
    "../data/reports/branch_summary.csv",
    index=False
)
```

agent summary

```python
agent_summary = (
    sales.groupby("agent_name")
    .agg(
        Phones_Sold=("items", "count"),
        Revenue=("sp", "sum"),
        Profit=("margin", "sum")
    )
    .sort_values("Phones_Sold", ascending=False)
    .reset_index()
)

agent_summary
```

```python
agent_summary.to_csv(
    "../data/reports/agent_summary.csv",
    index=False,
    encoding="utf-8-sig"
)
```

brand summary

```python
brand_summary.to_csv(
    "../data/reports/brand_summary.csv",
    index=False,
    encoding="utf-8-sig"
)
```
